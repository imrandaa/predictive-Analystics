# Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

# Load Dataset

```python
data = pd.read_excel("C:/Users/imran/OneDrive/Documents/sales_data.xlsx")
print(data)
```

# Convert Month into Number

```python
data["Month_No"] = range(1, len(data) + 1)
```

# Target

```python
X = data[["Month_No"]]
y = data["Sales"]
```

# Model

```python
model = LinearRegression()
model.fit(X, y)
```

# Predict next 3 month

```python
future = pd.DataFrame({"Month_No": [13, 14, 15]})
pred = model.predict(future)
```

```python
print("Predicted Sales")
print("Month 13:", round(pred[0], 2))
print("Month 14:", round(pred[1], 2))
print("Month 15:", round(pred[2], 2))
```

# Plot

```python
plt.scatter(data["Month_No"], data["Sales"], label="Historical Data")
plt.plot(data["Month_No"], model.predict(X), label="Regression Line")
plt.scatter(future["Month_No"], pred, marker="x", s=100, label="Prediction")

plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Sales Prediction")
plt.legend()
plt.show()
```
